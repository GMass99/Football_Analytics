```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
from PIL import Image
```

```python
df = pd.read_excel(r"C:\Users\George\Documents\Football Analytics\Data\Pukki_Sargent.xlsx")
```

```python
df.columns
```

```python
df.rename(columns={'SoT': 'Shots on Target', 'Ast' : 'Assists', 'G/Sh' : 'Goals per Shot'}, inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df = df.drop(columns=['xA'])
```

```python
# Parameters
params = list(df.columns)
params = params[1:]
params
```

```python
df
```

```python
# Range
ranges = []
a_values = []   
b_values = []

for x in params:
    a = min(df[params][x])
    a = a - (a*0.5)
    b = max(df[params][x])
    b = b + (b*0.15)
    #ranges.append((a,b))

for x in range(len(df['Unnamed: 0'])):
    if df['Unnamed: 0'][x] == 'Josh Sargent 23/24':
        a_values = df.iloc[x].values.tolist()
    if df['Unnamed: 0'][x] == 'Teemu Pukki 18/19':
        b_values = df.iloc[x].values.tolist()
```

```python
ranges = [(0.3,1.2),(2.2,4),(0.5,2.3),(0.1,1),(0.05,0.35),(0,0.3)]
ranges
```

```python
a_values = a_values[1:]
```

```python
b_values = b_values[1:]    
```

```python
values = [a_values,b_values]
values
```

```python
title = dict(
    title_name='Josh Sargent',
    title_color = '#9B3647',
    subtitle_name = '23/24',
    subtitle_color = '#9B3647',
    title_name_2='Teemu Pukki',
    title_color_2 = '#3282b8',
    subtitle_name_2 = '18/19',
    subtitle_color_2 = '#3282b8',
    title_fontsize = 18,
    subtitle_fontsize = 15
)

endnote = '@george__massey\nData from FBref'
```

```python
radar = Radar(background_color="#161616", patch_color='#28252C', label_fontsize=13.5, label_color='#F0FFF0', range_fontsize=9, range_color='#F0FFF0')

fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                           radar_color=['#9B3647','#3282b8'], alphas=[0.55,0.5], 
                           title=title, endnote=endnote, compare=True, image_coord=(0.458, 0.8, 0.11, 0.11),
                           image="C:/Users/George/Documents/Football Analytics/Team Logos/norwich city_logo.png" )

ax2 = fig.add_axes([.61,.7,.07,.11])
ax2.axis('off')
img = Image.open("C:/Users/George/Documents/Football Analytics/Images/pukki.png")
ax2.imshow(img)
ax3 = fig.add_axes([.34,.7,.07,.11])
ax3.axis('off')
img2 = Image.open("C:/Users/George/Documents/Football Analytics/Images/Sargent2.png")
ax3.imshow(img2)
fig.text(0.51, 0.945, 'Sargent vs Pukki', ha='center', fontsize=20, color='#F0FFF0', fontweight='bold')
fig.text(0.51, 0.92, 'All stats per 90 mins', ha='center', fontsize=14, color='#F0FFF0')

```
