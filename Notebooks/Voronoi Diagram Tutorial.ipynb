```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch, VerticalPitch
```

```python
barca_x=[93,96,99,102,104,95,86,79,89]
barca_y=[9,22,33,44,56,63,52,33,22]
barca = list(np.zeros((9,),dtype=int))

opp_x=[99,100,102,104,104,102,97,98,94,77]
opp_y=[14,22,33,43,46,55,37,27,21,45]
opp = list(np.ones((10,),dtype=int))

df = pd.DataFrame({'x':barca_x+opp_x, 'y':barca_y+opp_y, 'team':barca+opp})
```

```python
points = np.column_stack((df['x'], df['y']))
```

```python
fig, ax = plt.subplots(figsize=(13, 8.5))
fig.set_facecolor('#38383b')
ax.patch.set_facecolor('#38383b')

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#38383b', 
                     line_color='white', axis=True, half=True)
pitch.draw(ax=ax)

x = df['x']
y = df['y']
team1, team2 = pitch.voronoi(x,y,df.team)
t1 = pitch.polygon(team1, ax=ax, edgecolor='white', facecolor='red', alpha=0.5, lw = 3)
t2 = pitch.polygon(team2, ax=ax, edgecolor='white', facecolor='blue', alpha=0.5, lw = 3)
```
