```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
x = np.random.randint(1,101,500)
y = np.random.randint(1,101,500)
df = pd.DataFrame({'x':x, 'y':y})
```

```python
my_colormap = LinearSegmentedColormap.from_list('my_colormap', ['#e0aaff', '#10002b'], N=100)
```

```python
KDE = sns.kdeplot(
    x = df.x, y = df.y, data = df,
    cmap=my_colormap, shade=True, 
    shade_lowest=False, thresh=0.05, n_levels=100
)
```
