```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\George\Documents\Football Analytics\Data\tottenham prem xg.csv")
```

```python
df['Match'] = range(1, 39)
```

```python
df['xGSMA'] = df['xG'].rolling(window=3).mean()
```

```python
df['xGASMA'] = df['xGA'].rolling(window=3).mean()
```

```python
plt.plot(df['Match'], df['xGSMA'], label='xG', color = 'blue')
plt.plot(df['Match'], df['xGASMA'], label='xGA', color = 'red')
plt.xlabel('Match')
plt.ylabel('xG')
plt.title('Tottenham xG and xGA Moving Average')
plt.legend();
```
