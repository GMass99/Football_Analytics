```python
league_name = 'League One'
league_id = '108'
ccode = 'GBR'
season = '2023/2024'
round = '1'
```

```python
params = {
    'season': season,
    'ccode3': ccode,
    'id': league_id
}
response = requests.get('https://www.fotmob.com/api/leagues',params=params)
data = response.json()
```

```python
data = pd.DataFrame(data['matches']['allMatches'])
```

```python
df = data
```

```python
df = df[df['id'] != '4193691'].reset_index(drop=True)
df = df[df['id'] != '4202926'].reset_index(drop=True)
df = df[df['id'] != '4203025'].reset_index(drop=True)
df = df[df['id'] != '4203267'].reset_index(drop=True)
df = df[df['id'] != '4203250'].reset_index(drop=True)
```

```python
match_df = df
```

```python
match_df
```

```python
params = {
        'matchId': df['id'][0]
    }
response = requests.get('https://www.fotmob.com/api/matchDetails',params=params)
data = response.json()
```

```python
temp_list = []
counter = 0 
for id in match_df['id']:
    params = {
        'matchId': id
    }
    response = requests.get('https://www.fotmob.com/api/matchDetails',params=params)
    data = response.json()
    counter += 1
    print(counter)
    if data['header']['status']['cancelled'] == True:
        pass
    else:
        home_team_id = data['general']['homeTeam']['id']
        home_team_name = data['general']['homeTeam']['name']
        away_team_id = data['general']['awayTeam']['id']
        away_team_name = data['general']['awayTeam']['name']
        home_goals = data['header']['teams'][0]['score']
        away_goals = data['header']['teams'][1]['score']
        try:
            home_xG = data['content']['stats']['Periods']['All']['stats'][2]['stats'][1]['stats'][0]
            away_xG = data['content']['stats']['Periods']['All']['stats'][2]['stats'][1]['stats'][1]
            home_xGOT = data['content']['stats']['Periods']['All']['stats'][2]['stats'][-1]['stats'][0]
            away_xGOT = data['content']['stats']['Periods']['All']['stats'][2]['stats'][-1]['stats'][1]
        except:
            home_xG = home_goals
            away_xG = away_goals
            home_xGOT = home_goals
            away_xGOT = away_goals

        date = data['general']['matchTimeUTCDate']
        temp_list.append([id,date,home_team_name,home_team_id,home_xG,away_xG,home_xGOT,away_xGOT,home_goals,away_goals,'H'])
        temp_list.append([id,date,away_team_name,away_team_id,away_xG,home_xG,away_xGOT,home_xGOT,away_goals,home_goals,'A'])
```

```python
df = pd.DataFrame(temp_list,columns=['id','date','team_name','team_id','xG_for','xG_ag','xGOT_for','xGOT_ag','goals_for','goals_ag','venue'])
```
