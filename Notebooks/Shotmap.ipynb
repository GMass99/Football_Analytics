```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, FontManager
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\George\Documents\Football Analytics\Data\shotmaps.csv")
```

```python
df
```

```python
df_saved = df[df['outcome']=='Saved']
df_post = df[df['outcome']=='Post']
df_missed = df[df['outcome']=='Missed']
df_goal = df[df['outcome']=='Goal']
```

```python
fm_rubik = FontManager('https://raw.githubusercontent.com/google/fonts/main/ofl/'
                       'rubikmonoone/RubikMonoOne-Regular.ttf')
```

```python
pitch = VerticalPitch(half=True, goal_type = 'box', pad_bottom = 0.5)
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')
ax.invert_yaxis()

ax.scatter(df_saved.x,df_saved.y, c = 'blue', label = 'Saved', s = 100)
ax.scatter(df_post.x,df_post.y, c = 'orange', label = 'Post', s = 100)
ax.scatter(df_missed.x,df_missed.y, c = 'red', label = 'Missed', s = 100)
ax.scatter(df_goal.x,df_goal.y, c = 'green', label = 'Goal', s = 100)

legend = ax.legend(facecolor='#22312b', handlelength=5, 
                              edgecolor='none', loc='lower left')
legend.get_texts()[0].set_color('white')
legend.get_texts()[1].set_color('white')
legend.get_texts()[2].set_color('white')
legend.get_texts()[3].set_color('white')
for text in legend.get_texts():
    text.set_fontsize(20)

txt = ax.text(x=40, y=75, s='Barcelona shots\nversus Juventus',
              size=30,
              # here i am using a downloaded font from google fonts instead of passing a fontdict
              fontproperties=fm_rubik.prop, color=pitch.line_color,
              va='center', ha='center')
txt2 = ax.text(x=40, y=80, s='@george__massey',
              size=15,
              # here i am using a downloaded font from google fonts instead of passing a fontdict
              fontproperties=fm_rubik.prop, color=pitch.line_color,
              va='center', ha='center')
```
