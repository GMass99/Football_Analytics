```python
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\George\Documents\Football Analytics\Data\xgtutorial.csv")
```

```python
df
```

```python
a_xG = [0]
h_xG = [0]
a_min = [0]
h_min = [0]

# find home and away teams

hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]

for x in range(len(df['xG'])):
    if df['team'][x] == ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x] == hteam:
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
```

```python
def nums_cum_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]
a_cumulative = nums_cum_sum(a_xG)
h_cumulative = nums_cum_sum(h_xG)
```

```python
fig, ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Minute')
plt.ylabel('xG')

ax.step(x=a_min, y=a_cumulative, where = 'post')
ax.step(x=h_min, y=h_cumulative, where = 'post')
```
