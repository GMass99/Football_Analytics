```python
import re
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

```python
league = 'premier-league'
div = '1'
country = 'GB'
games = 38
```

```python
url = f'https://www.transfermarkt.co.uk/{league.lower()}/spieltagtabelle/wettbewerb/{country}{div}?saison_id=2023&spieltag=46'
```

```python
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
}
```

```python
response = requests.get(url, headers=headers)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
teams = []
list = soup.select('td[class="no-border-links hauptlink"]')
for y in range(len(list)):
    team = list[y].text.strip()
    teams.append(team)
```

```python
table = pd.DataFrame(teams)
```

```python
for x in range(1, games+1):
    url = f'https://www.transfermarkt.co.uk/{league.lower()}/spieltagtabelle/wettbewerb/{country}{div}?saison_id=2023&spieltag={x}'
    response = requests.get(url, headers=headers)
    response.status_code
    soup = BeautifulSoup(response.content, 'html.parser')
    list = soup.select('td[class="no-border-links hauptlink"]')
    teams = []
    for y in range(len(list)):
        team = list[y].text.strip()
        teams.append(team)
    df_temp = pd.DataFrame(teams)
    df_temp[f'Week {x}'] = range(1, len(df_temp) + 1)
    table = pd.merge(table, df_temp, on=0)
```
