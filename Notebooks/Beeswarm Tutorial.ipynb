```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\George\Documents\Football Analytics\Data\beeswarmTutorial.csv")
```

```python
df['per90'] = df['Prog']/df['90s']
```

```python
df = df[df['Pos'] != 'GK']
```

```python
df = df[df['90s'] >= 6.5]
```

```python
df.reset_index(drop=True, inplace=True) 
```

```python
df.head(20)
```

```python
background = '#313332'
text_colour = 'white'

fig, ax = plt.subplots(figsize=(12, 8))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

# set up base layer
mpl.rcParams['xtick.color'] = text_colour
mpl.rcParams['ytick.color'] = text_colour
spines = ['top', 'left', 'right'] 
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)
ax.spines['bottom'].set_color('white')

sns.swarmplot(x='per90', data=df, size=10, ax=ax, palette='viridis', zorder=1);
plt.annotate('Thiago Alcantara', xy=(9.879518, 0), xytext=(9, -0.002), fontsize=12, color='white');

plt.scatter(x=9.879518, y=0, c='red', s=100, zorder = 2);
plt.xlabel('Progressive Passes per 90', fontsize=14, color=text_colour);
plt.title('Beeswarm Plot for Prem Players Progressive Passes per 90', fontsize=16, color=text_colour);
```

```python
df2 = pd.read_csv(r"C:\Users\George\Documents\Football Analytics\Data\beeswarm2.csv")
```

```python
# split player names
df2['Player'] = df2['Player'].str.split('\\', expand=True)[0]
```

```python
df2 = df2[df2['Pos'] != 'GK']
df2 = df2[df2['90s'] >= 6.5]
df2.reset_index(drop=True, inplace=True)
```

```python
# make data per 90
df2['Prog'] = df2['Prog']/df2['90s']
df2['1/3'] = df2['1/3']/df2['90s']
df2['xA'] = df2['xA']/df2['90s']
df2['KP'] = df2['KP']/df2['90s']
df2['PPA'] = df2['PPA']/df2['90s']
```

```python
# create a list of metrics
metrics = ['Prog', '1/3', 'xA', 'Cmp%','KP', 'PPA']
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 10))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)
mpl.rcParams['xtick.color'] = text_colour
mpl.rcParams['ytick.color'] = text_colour

counter = 0
counter2 = 0
met_counter = 0

for i, ax in zip(df2['Player'], axes.flatten()):
    ax.set_facecolor(background)
    spines = ['top', 'bottom','left', 'right']
    for x in spines:
        if x in spines:
            ax.spines[x].set_visible(False)
    sns.swarmplot(x=metrics[met_counter], data=df2, ax=axes[counter,counter2], palette='viridis', zorder=1);
    ax.set_xlabel(f'{metrics[met_counter]}', fontsize=12, color=text_colour);

    for x in range(len(df2['Player'])):
        if df2['Player'][x] == 'Jack Grealish':
            ax.scatter(x=df2[metrics[met_counter]][x], y=0, c='red', s=100, zorder = 2);

        if df2['Player'][x] == 'James Maddison':
            ax.scatter(x=df2[metrics[met_counter]][x], y=0, c='white', s=100, zorder = 2);

    met_counter += 1
    if counter2 == 0:
        counter2 = 1
        continue
    if counter2 == 1:
        counter2 = 0
        counter += 1
plt.suptitle('Beeswarm Plot for Prem Players, James Maddison (White) vs Jack Grealish (Red)', fontsize=16, color=text_colour);
```
