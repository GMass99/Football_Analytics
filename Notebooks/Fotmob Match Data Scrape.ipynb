```python
import requests
import pandas as pd
```

```python
team_id = 10202
team_name = 'Nordsjælland'
league_name = 'Superligaen'
league_name2 = "Superligaen Championship Playoff"
```

```python
params = {
    'id':f'{team_id}',
    'ccode3': 'GBR'
}
response = requests.get('https://www.fotmob.com/api/teams',params=params)
data = response.json()
```

```python
id_list = []
for i in range(len(data['fixtures']['allFixtures']['fixtures'])):
    id_list.append(data['fixtures']['allFixtures']['fixtures'][i]['id'])
```

```python
df = pd.DataFrame({'home_team':[],'away_team':[],'team_xG':[],'opponent_xG':[]})
for i in id_list:
    params ={
        'matchId':f'{i}'
    }
    response = requests.get('https://www.fotmob.com/api/matchDetails',params=params)
    data = response.json()
    if data['general']['leagueName'] == league_name or data['general']['leagueName'] == league_name2:
        home_team = data['header']['teams'][0]['name']
        away_team = data['header']['teams'][1]['name']
        if home_team == team_name:
            team_xG = data['content']['stats']['Periods']['All']['stats'][0]['stats'][1]['stats'][0]
            opponent_xG = data['content']['stats']['Periods']['All']['stats'][0]['stats'][1]['stats'][1]
        else:
            team_xG = data['content']['stats']['Periods']['All']['stats'][0]['stats'][1]['stats'][1]
            opponent_xG = data['content']['stats']['Periods']['All']['stats'][0]['stats'][1]['stats'][0]
        df_temp = pd.DataFrame({'home_team':home_team,'away_team':away_team,'team_xG':team_xG,'opponent_xG':opponent_xG},index=[0])
        df = pd.concat([df,df_temp])
    else:
        pass
```

```python
df['Match'] = range(1,len(df)+1)
```
